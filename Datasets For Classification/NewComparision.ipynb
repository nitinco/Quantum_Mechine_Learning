```python
import pandas as pd
import glob
import os

csv_files = glob.glob("**/dataset*.csv", recursive=True)
print(csv_files)

combined = {}

for file_path in csv_files:
    df=pd.read_csv(file_path,header=None,index_col=0)
    dataset_name = os.path.splitext(os.path.basename(file_path))[0]
    combined[dataset_name] = df[1]

final_df = pd.DataFrame(combined)
final_df.to_csv("AllMergedScores.csv")
```
