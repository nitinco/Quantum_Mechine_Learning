```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import pandas as pd
```

```python
#load Data
data = pd.read_csv('Book1.csv')  # from project root
print(data.to_string())
data.plot()
plt.show()
```

```python
plt.xlabel('Study Time')
plt.ylabel('Result_Score')
plt.scatter(data.Study_Time,data.Result_score, color='red',marker='_')
```

```python
modal=LinearRegression()
modal.fit(data[['Study_Time']],data.Result_score)
```

```python
modal.predict([[5]])
```

```python
print(modal.coef_)
print(modal.intercept_)
```

```python
#to understand how mx+c working here m=coef,x=value to be predict and c= interseptor point
7.8*5+20.66
```
